```python
import json
import matplotlib.pyplot as plt

def visualize_json(json_file, keys):
    # Read the JSON file
    with open(json_file, 'r') as f:
        data = [json.loads(line) for line in f]

    data = data[:-1]
    print(data[-1])
    # Extract values of the given keys into lists
    values = {key: [entry[key] for entry in data] for key in keys}

    # Plotting
    fig, ax1 = plt.subplots()

    color = 'tab:red'
    ax1.set_xlabel('Iteration')
    ax1.set_ylabel(keys[0], color=color)
    ax1.plot(values[keys[0]], color=color, label=keys[0])
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'tab:blue'
    ax2.set_ylabel(keys[1], color=color)  # we already handled the x-label with ax1
    ax2.plot(values[keys[1]], color=color, label=keys[1])
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.title(f'Visualization of "{keys[0]}" and "{keys[1]}" from JSON file')
    plt.grid(True)
    fig.legend()
    plt.show()

# Example usage
if __name__ == "__main__":
    # json_file = input("Enter the name of the JSON file: ")
    json_file = '/home/hhiromasa/code/Uni-Perceiver/asset/FT_glue_fin/metrics.json'
    # keys = input("Enter the dictionary keys to visualize (separated by space): ").split()
    keys = ['total_grad','QQP/Accuracy']
    visualize_json(json_file, keys)
```
